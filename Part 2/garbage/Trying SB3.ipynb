```python
from stable_baselines3 import A2C, PPO
from stable_baselines3.common.env_util import make_vec_env
from stable_baselines3.common.evaluation import evaluate_policy

import time

from Callback import SaveOnBestTrainingRewardCallback

# Parallel environments
vec_env = make_vec_env("CartPole-v1", n_envs=64)


model = PPO("MlpPolicy", vec_env, verbose=1)

start = time.time()
model.learn(total_timesteps=10000)
model.save("a2c_cartpole")

del model # remove to demonstrate saving and loading

model = PPO.load("a2c_cartpole")

print(time.time() - start)
print("evaluation...")
test_env = make_vec_env("CartPole-v1", n_envs=1)
test_env.reset()

mean_reard, std_reward = evaluate_policy(model, test_env, n_eval_episodes=50, render=False)
print(mean_reard, std_reward)

```
