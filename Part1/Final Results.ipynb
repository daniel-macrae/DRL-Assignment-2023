```python
import numpy as np
import matplotlib.pyplot as plt
import os

avgArr = np.zeros(300)
window_size = 10
#episodes = np.arange(10, 5000, 10)

average_rewards = []
max_rewards = []

folder = "FinalResults"
for file in os.listdir(folder):
    path = os.path.join(folder, file)
    arr = np.load(path)
    avgArr += arr
    plt.plot(np.arange(0, 3000, 10), arr, linewidth = 1)

    average_rewards.append(sum(arr)/len(arr))

    maxReward = 0

    for i in range(len(arr) - window_size + 1):
        windowAvgReward = sum(arr[i : i + window_size]) / window_size
        #average_rewards.append(avgReward)
        if windowAvgReward > maxReward:  maxReward = windowAvgReward
    max_rewards.append(maxReward)

avgArr /= len(os.listdir(folder))

# ploting the models training individually
plt.xlabel("Training Episodes")
plt.ylabel("Reward")
plt.title("Rewards over testing episodes")
plt.show()

# ploting the average rewards (on the testing episodes) over all 5 models
plt.plot(np.arange(0, 3000, 10), avgArr, linewidth = 1)
plt.xlabel("Training Episodes")
plt.ylabel("Reward")
plt.title("Rewards averaged over 5 models")
plt.show()


print("averaged over each model's training:")
print("                                Mean           STD")
print(f"average episode reward    ||   {np.mean(average_rewards):.3f}   ||    {np.std(average_rewards):.3f}")
print(f"maximum episode reward    ||   {np.mean(max_rewards):.3f}   ||    {np.std(max_rewards):.3f}")
```
