```python
import numpy as np
import matplotlib.pyplot as plt
import os

avgArr = np.zeros(499)
episodes = np.arange(10, 5000, 10)

folder = "Results2"
for file in os.listdir(folder):
    path = os.path.join(folder, file)
    arr = np.load(path)
    avgArr += arr


avgArr /= len(os.listdir(folder))
plt.plot(episodes, avgArr)
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Reward curve averaged over 5 runs")
plt.show()
```
