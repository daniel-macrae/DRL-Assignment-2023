```python
import numpy as np
import matplotlib.pyplot as plt
import os

avgArr = np.zeros(300)
#episodes = np.arange(10, 5000, 10)

folder = "FinalResults"
for file in os.listdir(folder):
    path = os.path.join(folder, file)
    arr = np.load(path)
    avgArr += arr

avgArr /= len(os.listdir(folder))
plt.plot(avgArr)
plt.xlabel("Episode")
plt.ylabel("Reward")
plt.title("Reward curve averaged over 5 runs")
plt.show()
```
